```python
import seaborn as sns
from sklearn.preprocessing import LabelEncoder 
print('Version Seaborn : ' + sns.__version__)
tips = sns.load_dataset('tips') # Chargement des donn√©es tips
ec = LabelEncoder()
print('Feature Engineering Data Tips')
tips["sex"] = ec.fit_transform(tips['sex'].astype('str'))
tips["smoker"] = ec.fit_transform(tips['smoker'].astype('str'))
tips["day"] = ec.fit_transform(tips['day'].astype('str'))
tips["time"] = ec.fit_transform(tips['time'].astype('str'))
print('Normalization Data Tips')
tips_normalized = (tips-tips.min())/(tips.max()-tips.min())
# print('Standardization Data Tips')
tips_normalized.head(10)
```
